## Solo necesitamos los canales de colores RGB, quitamos la capa 4 de los colores

```python
img.shape
```

```python
img_rgb = img[:,:,:3]
```

```python
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

## Invertimos los colores

```python
img_inv = 1-img_rgb
```

```python
plt.imshow(img_inv)
```

## Si lo pasamos a escala de grises: (sacar promedio del rgb)

```python
img_gray = img_inv.mean(axis=2)
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
img_pil = Image.fromarray(np.uint8(img_gray*255))
```

```python
type(img_pil)
```

```python
print(tesserocr.image_to_text(img_pil, lang="spa"))
```

### Intentemos con imagen invertida pero conservando el rgb

```python
img_pil_inv = Image.fromarray(np.uint8(img_inv*255))
print(tesserocr.image_to_text(img_pil_inv,lang="spa"))
```
